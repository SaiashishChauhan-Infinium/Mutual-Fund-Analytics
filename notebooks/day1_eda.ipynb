Validate AMFI Codes

```python
fund_master = pd.read_csv(r"C:\Users\Saiashish\Desktop\Mutual-Fund-Analytics\data\raw\01_fund_master.csv")

nav_history = pd.read_csv(r"C:\Users\Saiashish\Desktop\Mutual-Fund-Analytics\data\raw\02_nav_history.csv")
```

```python
print("Fund Master Records:", fund_master.shape[0])

print("NAV History Records:", nav_history.shape[0])
```

```python
print(fund_master.columns)

print(nav_history.columns)
```

```python
master_codes = set(fund_master["amfi_code"])

nav_codes = set(nav_history["amfi_code"])
```

```python
missing_codes = master_codes - nav_codes
```

```python
print("Total Codes in Fund Master:",len(master_codes))

print("Total Codes in NAV History:",len(nav_codes))

print("Missing Codes:",len(missing_codes))
```

```python
# Missing values
print(fund_master.isnull().sum().sum())
print(nav_history.isnull().sum().sum())

# Duplicate rows
print(fund_master.duplicated().sum())
print(nav_history.duplicated().sum())

# Missing AMFI codes
print(len(missing_codes))
```

```python
if len(missing_codes) > 0:
    print(list(missing_codes)[:20])
else:
    print("All AMFI codes validated successfully.")
```
